```python
#Think of yourself, currently working as a Business analyst in one of the top sports company. 
#The senior management team has asked you to come up with metrics with which they can gauge which team will win the upcoming La Liga cup (Football tournament).
#The data set contains information on all the teams so far participated in all the past tournaments. It has data about how many goals each team scored, conceded, how many times they came within first 6 positions, how many seasons they have qualified, their best position in the past etc.

#Question 1 : Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.

import pandas as pd
import numpy as np

teams = pd.read_csv('Laliga.csv')
teams_df = teams.replace("-","0")
teams_df
```

```python
#Question 2 : Print all the teams which have started playing between 1930-1980

teams_df_after30 = teams_df.loc[teams_df["Debut"] >="1930",:]
print("team playing after 1930 & before 1980\n")
teams_df_after30_before80 = teams_df_after30.loc[teams_df_after30["Since/LastApp"]<="1980",:]
teams_df_after30_before80
```

```python
#Question 3 : Print the list of teams which came Top 5 in terms of points
#Understand the data types & convert to integers to perform mathematical operations
print("Teams data frame data types\n",teams_df.dtypes)
teams_df['Points'] = teams_df.Points.astype('int')
top5teams_df = teams_df.sort_values(by=["Points"],ascending = False)
print("Top 5 teams in terms of points\n",top5teams_df.head(5))
```

```python
#Question 4 : Write a function with name â€œGoal_diff_countâ€ which should return all the teams with their Goal Differences. 
#Using the same function, find the team which has maximum and minimum goal difference.

#type casting GoalsFor & GoalsAgainst columns into type int
teams_df['GoalsFor'] = teams_df.GoalsFor.astype('int')
teams_df['GoalsAgainst'] = teams_df.GoalsAgainst.astype('int')

#Add a new column to data frame called "GoalDifference" & assign 0 for ref
teams_df["GoalDifference"] = 0

#Goal differemce is calculated as the difference between goalsFor and goals against 
rowCount  = len(teams_df.index)
print("Row count:",rowCount)

def Goal_diff_count():
    for i in range(0,rowCount,1):
        teams_df.loc[i,"GoalDifference"] = teams_df.loc[i,"GoalsFor"] - teams_df.loc[i,"GoalsAgainst"]
    return

Goal_diff_count()

goalDiff_df = teams_df.sort_values(by=['GoalDifference'],ascending= False)
print("Team with max goal difference:",goalDiff_df.iloc[0,1])
print("Team with min goal difference:",goalDiff_df.iloc[rowCount-1 ,1])
goalDiff_df
```

```python
#Question 5 :Create a new column with name â€œWinning Percentâ€ and append it to the data set.
#Percentage of Winning = (GamesWon / GamesPlayed)*100.
#If there are any numerical error, replace it with 0%
#Print the top 5 teams which has the highest Winning percentage

teams_df['GamesWon'] = teams_df.GamesWon.astype('int')
teams_df['GamesPlayed'] = teams_df.GamesPlayed.astype('int')
teams_df['WinningPercent'] = ((teams_df['GamesWon']) / (teams_df['GamesPlayed'])) * 100
print("Nan values : ",teams_df['WinningPercent'].isnull().sum())
teams_df['WinningPercent'].fillna("0",inplace=True)
print("Nan values : ",teams_df['WinningPercent'].isnull().sum())
teams_df['WinningPercent'] = teams_df.WinningPercent.astype('int')
top5winningteams = teams_df.sort_values(by=['WinningPercent'],ascending= False)
top5winningteams.head(5)
```

```python
#Question 6 : Group teams based on their â€œBest positionâ€ and print the sum of their points for all positions
bestpositeamAndSum = teams_df.groupby("BestPosition").Points.sum()
bestpositeamAndSum
```
